# Usage Example for linearPlot.py

```python
%matplotlib inline

from linearPlot import *

# Initalize list of x-values
x = [5, 10, 13, 16, 20]
x = [1/p for p in x]

# Initialize x uncertainty
xe = [0.01]*5

# Initialize y-values
y = [5.08, 2.19, 1.67, 1.45, 1.05]

# Initialize y uncertainty
ye = [0.13]*5

linearPlot(x, xe, y, ye, "Frequency vs Wavelength", "1/Frequency", "Wavelength")

# Show Plots
plt.show()
```
